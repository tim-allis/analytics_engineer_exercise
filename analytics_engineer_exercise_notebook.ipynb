# list all tables and views in duckdb file (includes dbt models after dbt run)

```python
conn = duckdb.connect(duckdb_file)

# list tables, columns, and column data types
sql = """
    SELECT 
        table_name,
        table_schema,
        table_type
    FROM 
        information_schema.tables    
    ORDER BY 
        table_schema, 
        table_name;
"""

# Execute the query and fetch results
result = conn.execute(sql)
table_info = result.fetchall()

for row in table_info:
    table_name = row[0]
    table_schema = row[1]
    table_type = row[2]
    
    print(f"Table Name: {table_name}")
    print(f"Schema: {table_schema}")
    print(f"Type: {table_type}")
    print()  
```

# review raw source tables 

```python
conn = duckdb.connect(duckdb_file)

# list tables, columns, and column data types
sql = """
    SELECT 
        t.table_name,
        t.table_schema,
        t.table_type,
        c.column_name,
        c.data_type
    FROM 
        information_schema.tables AS t
        JOIN information_schema.columns AS c
        ON t.table_name = c.table_name
        AND t.table_schema = c.table_schema
    WHERE
        t.table_schema = 'main'
    ORDER BY 
        t.table_schema, t.table_name, c.ordinal_position;
"""

# Execute the query and fetch results
result = conn.execute(sql)
table_info = result.fetchall()

# Initialize variables to track the current table
current_table = None

# Print the results grouped by table and schema
for row in table_info:
    table_name, table_schema, table_type, column_name, data_type = row
    
    # Check if we need to print a new table header
    if current_table != (table_name, table_schema):
        if current_table is not None:
            print()  # Add a blank line between tables
        print(f"Table Name: {table_name}")
        print(f"Schema: {table_schema}")
        print(f"Type: {table_type}")
        current_table = (table_name, table_schema)
    
    # Print column details
    print(f"  Column Name: {column_name}")
    print(f"  Data Type: {data_type}")

conn.close()
```

# extend sql for intm_pricing__tariffs dbt model

```python
conn = duckdb.connect(duckdb_file)

sql = '''
    WITH extract_fields AS (
        SELECT
            rule_id,        
            tariff->>'$.price' AS price,        
            tariff->>'$.size' AS duration,        
            tariff->>'$.applies_on_days[*]' AS applies_on_days,
            tariff->>'$.applies_from' AS applies_from,  
            tariff->>'$.applies_to' AS applies_to,
            tariff->>'$.interval' AS interval,
            tariff->>'$.repeat' AS repeat,
            tariff->>'$.applies_between' AS applies_between
        FROM
            "pricing"."main_base"."base_pricing__rules"
    ),

    add_valid_from_to AS (
        SELECT 
            * EXCLUDE (applies_between, duration),
            TRY_CAST(duration AS INTEGER) AS duration, 
            CAST(SUBSTRING(applies_between, 4, 10) AS DATE) AS tariff_valid_from,
            CAST(SUBSTRING(applies_between, 17, 10) AS DATE) AS tariff_valid_to        
        FROM 
            extract_fields        
    ),

    extract_applies_on_days AS (
        SELECT
            *,            
        FROM  
            add_valid_from_to, UNNEST(applies_on_days) AS rule_applies_on
    )

    SELECT 
        p.* EXCLUDE(applies_on_days, applies_on_days_1),
        applies_on_days_1 AS applies_on_day,
        day_of_week_name
    FROM   
        extract_applies_on_days AS p
        INNER JOIN pricing.main_seeds.day_of_week_map AS s
        ON p.applies_on_days_1 = s.day_of_week_index
    WHERE
        rule_id = '883511'
    ORDER BY 
        duration,
        applies_on_day
        
'''

df_i_p_tariffs_extend = conn.execute(sql).fetchdf()
conn.close()
print(f"intm pricing tariffs model shape: {df_i_p_tariffs_extend.shape}\n")
df_i_p_tariffs_extend.head(df_i_p_tariffs_extend.shape[0])
df_i_p_tariffs_extend.head(21)
```

# sql for intm_pricing__prices dbt model

```python
# first check if there are locations without a null deleted_at value - this assumes that a 'null deleted_at' row = current applicable rule_id
# zero rows are returned so determined that all location_id values have a 'current' applicable rule_id

conn = duckdb.connect(duckdb_file)

sql = '''    
    WITH null_deleted_at AS (
        SELECT
            price_id,
            location_id,
            pricing_mode,
            created_at,                        
            deleted_at,
            rule_id
        FROM
            "pricing"."main_base"."base_pricing__prices"
        WHERE
            deleted_at IS NULL
    )
    
    SELECT DISTINCT
        p.location_id,        
    FROM  
        null_deleted_at AS n
        LEFT JOIN  "pricing"."main_base"."base_pricing__prices" AS p
        ON n.location_id = p.location_id
    WHERE
        n.location_id IS NULL

'''

df_chk_intm_pricing_prices = conn.execute(sql).fetchdf()
conn.close()
print(f"intm pricing prices model shape: {df_chk_intm_pricing_prices.shape}\n")
df_chk_intm_pricing_prices.head()
```

# Q1 - _current_ price that would be charged on different days of the week for parking sessions of various durations, at a specific location

```python
# set values of required variables
# I have not added error handling to check the value entered for these variables

v_location_descriptions = ['503495']
v_days_of_week = ['Monday']
v_durations = [1]
v_deleted_at = 'IS NULL' # deleted at is null equals current price
v_now = datetime.now().strftime('%Y-%m-%d')

v_location_descriptions_str = ', '.join(f"'{location}'" for location in v_location_descriptions)
v_days_of_week_str = ', '.join(f"'{day}'" for day in v_days_of_week)
v_durations_str = ', '.join(f"{duration}" for duration in v_durations)


conn = duckdb.connect(duckdb_file)

sql = f'''   
    SELECT        
        location_description,
        p.location_id,
        rule_id,
        applies_on_day_name,
        duration,
        tariff_valid_from,
        tariff_valid_to,
        created_at,
        deleted_at,
        price
    FROM
        "pricing"."main_mart"."fact_prices_tariffs" AS p  
        INNER JOIN "pricing"."main_mart"."dim_locations" AS l  
        ON p.location_id = l.location_id
    WHERE 
        l.location_description IN ({v_location_descriptions_str})
        AND applies_on_day_name IN ({v_days_of_week_str})
        AND duration IN ({v_durations_str})
        AND deleted_at {v_deleted_at}       
        AND tariff_valid_to > '{v_now}'
'''

df_q1 = conn.execute(sql).fetchdf()
conn.close()
print(f"Q1 shape: {df_q1.shape}\n")
df_q1.head()
```

# Q2 - how the price for a one hour parking session at a specific location has changed over time

```python
# set values of required variables
# I have not added error handling to check the value entered for these variables

v_location_descriptions = ['503495']
v_days_of_week = ['Monday']
v_durations = [1]

v_location_descriptions_str = ', '.join(f"'{location}'" for location in v_location_descriptions)
v_days_of_week_str = ', '.join(f"'{day}'" for day in v_days_of_week)
v_durations_str = ', '.join(f"{duration}" for duration in v_durations)


conn = duckdb.connect(duckdb_file)

sql = f'''   
    SELECT        
        location_description,        
        created_at,        
        price
    FROM
        "pricing"."main_mart"."fact_prices_tariffs" AS p  
        INNER JOIN "pricing"."main_mart"."dim_locations" AS l  
        ON p.location_id = l.location_id
    WHERE 
        l.location_description IN ({v_location_descriptions_str})
        AND applies_on_day_name IN ({v_days_of_week_str})
        AND duration IN ({v_durations_str})
    ORDER BY 
        created_at    
'''

df_q2 = conn.execute(sql).fetchdf()
conn.close()
print(f"Q2 shape: {df_q2.shape}\n")

plt.figure(figsize=(10, 6))
plt.plot(df_q2['created_at'], df_q2['price'], marker='o', linestyle='-', color='b')
plt.title('Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
